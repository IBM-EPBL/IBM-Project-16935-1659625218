Importing req. lib.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

Image processiong

```python
# Create a image

img1 = np.zeros((400,600,3),np.uint8)
plt.imshow(img1)
```

```python
# Draw a circle

circle = cv2.circle(img1, (300,200), 50, (255,0,0), -1)   # (0,0,0)--->(R,G,B)
plt.imshow(img1)
```

```python
# Drawing rectangle

rectangle = cv2.rectangle(img1,(200,100),(400,300),(0,255,0),6)
plt.imshow(img1)
```

```python
# Drawing line

line1 = cv2.line(img1,(200,100),(400,300),(0,0,255),4)
line2 = cv2.line(img1,(200,300),(400,100),(0,0,255),4)
plt.imshow(img1)
```

```python
#text on image

text = cv2.putText(img1, 'openCV', (200,50), cv2.FONT_HERSHEY_SIMPLEX, 2, (255,255,255),5)
plt.imshow(img1)
```

```python
# Reading the image

img = cv2.imread('/content/download.jpg',1)
plt.imshow(img)
```

```python
# Convert BGR to RGB

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
# Convert BGR to Gray

img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray)
```

```python
# Finding shape

img_rgb.shape
```

```python
# Resize the image

resize = cv2.resize(img_rgb,(500,1000))
print(resize.shape)
plt.imshow(resize)
```

```python
# Edge Detection

edge = cv2.Canny(img_rgb,100,200)
plt.imshow(edge)
```

```python
 # Blur image

r = resize[130:370,150:300]
blur = cv2.GaussianBlur(r,(13,13),cv2.BORDER_DEFAULT)
plt.imshow(resize)
plt.imshow(blur)
```
